```python
import requests
import time
import pandas as pd
import numpy as np
import json
```

```python
def getlatlongname(postalcode):
    url = 'https://nominatim.openstreetmap.org/search?postalcode='+postalcode+'&format=json'
    r = requests.get(url)
    
    if len(r.json())>0:
        lat = r.json()[0]["lat"]
        lon = r.json()[0]["lon"]
        address = r.json()[0]["display_name"]
    
        return json.dumps(
            {'address': address,
             'lat': lat,
             'lon': lon})
    
    else:
        return json.dumps(
            {'address': np.nan,
             'lat': np.nan,
             'lon': np.nan})
```

```python
data = pd.read_csv("data/oncologists_inbound.csv") 
```

```python
data = data.reindex(columns = data.columns.tolist() + ['n1_lat','n1_lon','n1_address','n2_lat','n2_lon','n2_address'])
data
```

```python
for i, row in data.iterrows():
    
    time.sleep(0.25)
    n1_loc = getlatlongname(str(data.iloc[i].node_1_zip_code))
    n1_loc_json = json.loads(n1_loc)
    data.at[i,'n1_lat'] = n1_loc_json["lat"]
    data.at[i,'n1_lon'] = n1_loc_json["lon"]
    data.at[i,'n1_address'] = n1_loc_json["address"]
    
    n2_loc = getlatlongname(str(data.iloc[i].node_2_zip_code))
    n2_loc_json = json.loads(n2_loc)
    data.at[i,'n2_lat'] = n2_loc_json["lat"]
    data.at[i,'n2_lon'] = n2_loc_json["lon"]
    data.at[i,'n2_address'] = n2_loc_json["address"]
    
    print(data.iloc[i])
    print(str(round((i*100/len(data)),2))+"%")
```
