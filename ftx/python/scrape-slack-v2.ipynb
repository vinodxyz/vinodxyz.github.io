```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.common.exceptions import NoSuchElementException
from selenium.common.exceptions import StaleElementReferenceException

import pandas as pd 
import numpy as np
import time
import re
```

```python
url = "https://razorpay.slack.com/stats#members"
driver = webdriver.Firefox(executable_path=r'/Users/vinodkumarm/Documents/GitHub/vinodxyz.github.io/ftx/python/geckodriver')
driver.get(url)
```

```python
df = pd.DataFrame(columns=['no','name','role','url'])

x=0
i=0
totalcount = 0

driver.execute_script("document.getElementsByClassName('ent_data_table__scrollable ent_data_table__scrollable--width_0')[0].scrollTo(0, arguments[0]);",0)

while(totalcount<12000):
    
    ignored_exceptions=(NoSuchElementException,StaleElementReferenceException,)
    #items = WebDriverWait(driver,2,ignored_exceptions=ignored_exceptions).until(driver.find_elements_by_class_name('ent_data_table__row'))
    items = driver.find_elements_by_class_name('ent_data_table__row')
    item = items[i]

    user_name = ""
    user_role = ""
    user_avatar = ""

    if len(item.find_elements_by_class_name('ent_data_table__primary_text')) > 0:
        user_name = str(item.find_elements_by_class_name('ent_data_table__primary_text')[0].get_attribute("innerText")).replace(".", " ").replace("\n", "").replace("\t", "")

    if len(item.find_elements_by_class_name('ent_data_table__secondary_text')) > 0:
        user_role = item.find_elements_by_class_name('ent_data_table__secondary_text')[0].get_attribute("innerText")
        user_role = user_role.split("\n")[0]

    if len(item.find_elements_by_class_name('ent_avatar')) > 0:
        user_avatar = str(item.find_elements_by_class_name('ent_avatar')[0].get_attribute("src"))
    
    if len(items[i].get_attribute("style")) > 0:
        translatedby = items[i].get_attribute("style")
        yvalue = int(re.search('translateY\((.*)px', translatedby)[1])
        user_no = int(yvalue/60)
    

    print(str(user_no)+" : "+user_name+" /"+str(int(totalcount)))
    df.loc[len(df)] = [user_no,user_name,user_role,user_avatar]
    
    if i<17:
        i = i+1
    else:
        time.sleep(1)
        driver.implicitly_wait(1)
        i=0
        x = x + 60
        driver.execute_script("document.getElementsByClassName('ent_data_table__scrollable ent_data_table__scrollable--width_0')[0].scrollTo(0, arguments[0]);",x)
        
    totalcount = totalcount + 1

```

```python
df.drop_duplicates()
df.to_csv("user_list_permu.csv")
driver.quit()
```
